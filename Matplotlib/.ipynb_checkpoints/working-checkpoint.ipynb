```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

x2 = np.arange(0,4.5,0.5)
plt.plot(x,y, marker="o", color="red", label="Y is 2x"
         , markeredgecolor="b", markerfacecolor="b",linestyle="--")
plt.title("Testing")
plt.xlabel("x Axis")
plt.ylabel("y label")
plt.plot(x2[:6], x2[:6]**2, color="blue", label="x2^2")
plt.plot(x2[5:],x2[5:]**2, linestyle="--", color="blue")

# plt.xticks(range(1,5,1))
# plt.yticks([0,2,4,6,8])

plt.savefig("mygraphsave.png", dpi=300)

plt.legend()
plt.show()
```
